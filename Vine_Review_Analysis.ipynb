```python
import pandas as pd
```

```python
vine_table = pd.read_csv('vine_table.csv')  
df = pd.DataFrame(vine_table)
```

```python
df.head(5)
```

```python
total_votes_df = df.loc[df["total_votes"]>= 20]
total_votes_df
```

```python
helpful_votes_df = total_votes_df.loc[total_votes_df["helpful_votes"]/total_votes_df["total_votes"] >= 0.5]
helpful_votes_df
```

```python
vine_program_paid_df = helpful_votes_df.loc[helpful_votes_df["vine"] =="Y"]
vine_program_paid_df
```

```python
vine_program_unpaid_df = helpful_votes_df.loc[helpful_votes_df["vine"] =="N"]
vine_program_unpaid_df
```

```python
#Determine the total number of reviews, 
#and the percentage of 5-star reviews for the two types of review (paid vs unpaid).

total_reviews = helpful_votes_df["vine"].count()
total_reviews
```

```python
#the number of 5-star reviews, 
five_star_reviews = len(helpful_votes_df.loc[helpful_votes_df["star_rating"] == 5.0])
five_star_reviews
```

```python
#five_star reviews for paid vine program
five_star_paid = len(vine_program_paid_df.loc[vine_program_paid_df["star_rating"] == 5.0])
five_star_paid
```

```python
#five_star reviews for unpaid vine program
five_star_unpaid = len(vine_program_unpaid_df.loc[vine_program_unpaid_df["star_rating"] == 5.0])
five_star_unpaid
```

```python
#percentage of five_star reviews for paid and unpaid vine program 
paid_five_star_percentage = (five_star_paid/five_star_reviews) * 100
unpaid_five_star_percentage = (five_star_unpaid/five_star_reviews) * 100

print(paid_five_star_percentage)
print(unpaid_five_star_percentage)
```

```python
#percentage of five_star reviews for paid and unpaid vine program over total reviews

total_paid_five_star_percentage = (five_star_paid/total_reviews) * 100
total_unpaid_five_star_percentage = (five_star_unpaid/total_reviews) * 100

print(total_paid_five_star_percentage)
print(total_unpaid_five_star_percentage)
```

```python
#percentage of 5 stars/vine users and 5star/non users
vine_user_percent=(five_star_paid/len(vine_program_paid_df["vine"]))*100
print(vine_user_percent)

vine_non_user_percent=(five_star_unpaid/len(vine_program_unpaid_df["vine"]))*100
vine_non_user_percent
```
